```python
import voila
import ipywidgets as widgets
from IPython.display import display
import time
from threading import Timer

# Countdown timer class to handle the timing logic
class CountdownTimer:
    def __init__(self):
        self.time_left = 0
        self.interval = 1
        self.timer = None
        self.running = False
    
    def start_timer(self):
        if not self.running:
            self.running = True
            self._run_timer()
    
    def _run_timer(self):
        if self.time_left > 0 and self.running:
            self.time_left -= 1
            timer_display.value = str(self.time_left)
            self.timer = Timer(self.interval, self._run_timer)
            self.timer.start()
        else:
            self.running = False
    
    def pause_timer(self):
        if self.running:
            self.running = False
            if self.timer:
                self.timer.cancel()
    
    def reset_timer(self, time_left, interval):
        self.time_left = time_left
        self.interval = interval
        timer_display.value = str(self.time_left)
        if self.running:
            self.pause_timer()
            self.start_timer()

# Initialize countdown timer
countdown = CountdownTimer()

# Slider to select the starting number
starting_number_slider = widgets.IntSlider(
    value=60,
    min=1,
    max=600,
    step=1,
    description='Starting Number:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

# Slider to select the interval in seconds
interval_slider = widgets.IntSlider(
    value=1,
    min=1,
    max=300,
    step=1,
    description='Interval (seconds):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

# Display for the countdown timer
timer_display = widgets.Label(value="60", style={'description_width': 'initial', 'font_size':'72px', 'font_weight':'bold'},
                              layout=widgets.Layout(height='200px', display='flex', align_items='center'))

# Start button
start_button = widgets.Button(description="Start", layout=widgets.Layout(display='flex', justify_content='center', width='50%'))

# Pause/Resume button
pause_resume_button = widgets.Button(description="Pause", layout=widgets.Layout(display='flex', justify_content='center', width='50%'))

# Function to start the countdown
def on_start_button_click(b):
    countdown.reset_timer(starting_number_slider.value, interval_slider.value)
    countdown.start_timer()
    pause_resume_button.description = "Pause"  # Ensure button says "Pause"

# Function to pause or resume the countdown
def on_pause_resume_button_click(b):
    if countdown.running:
        countdown.pause_timer()
        pause_resume_button.description = "Resume"
    else:
        countdown.start_timer()
        pause_resume_button.description = "Pause"

# Link the buttons to their functions
start_button.on_click(on_start_button_click)
pause_resume_button.on_click(on_pause_resume_button_click)

# Layout the widgets vertically
app_layout = widgets.VBox([
    starting_number_slider,
    interval_slider,
    start_button,
    timer_display,
    pause_resume_button
], layout=widgets.Layout(width='100%', display='flex', align_items='center'))

# Display the entire application
display(app_layout)
```
